```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

from sklearn import preprocessing
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
%matplotlib inline
```

### 1. Use the bank.csv to explore the data. Observe the features: are they numbers? Are they strings? Are they binary? Are they continuous?

```python
data = pd.read_table('bank.csv', sep=';')
data.head()
```

```python
data.describe()
```

Age, balance, duration, campaign, pdays, previous, day are all continuous values. Job, marital, education, contract, month, poutcome are all categorical strings. Binary variables are default, housing, loan, and y. 

### 2. Learn about label encoders here: http://scikit- learn.org/stable/modules/preprocessing.html#encoding-categorical-features and transform the features to numerical features

```python
data_clean = data

new_label_vars = ["marital", "job", "education", "contact", "month", "poutcome", "default", "housing", "loan", "y"]

for i in new_label_vars:
    le = preprocessing.LabelEncoder()
    le.fit(data_clean[i].unique())
    newvarname = "new_" + str(i)
    data_clean[newvarname] = le.transform(data_clean[i]).tolist()

data_clean.drop(new_label_vars, inplace=True, axis=1)
```

### 3. Build a simple decision tree model to predict the classification goal. 4. Evaluate the result of the classification with cross-validation.

```python
feature_names = data_clean.columns.tolist()
feature_names.remove("new_y")

features = data_clean[feature_names].values
target = data_clean['new_y'].values
```

```python
# Cross validation function
from sklearn.cross_validation import KFold

def cross_validate(X, y, classifier, k_fold) :

    # derive a set of (random) training and testing indices
    k_fold_indices = KFold(len(X), n_folds=k_fold,
                           shuffle=True, random_state=0)

    k_score_total = 0
    # for each training and testing slices run the classifier, and score the results
    for train_slice, test_slice in k_fold_indices :

        model = classifier(X[ train_slice  ],
                         y[ train_slice  ])

        k_score = model.score(X[ test_slice ],
                              y[ test_slice ])

        k_score_total += k_score

    # return the average accuracy
    return k_score_total/k_fold
```

```python
model = tree.DecisionTreeClassifier(random_state=0).fit
cross_validate(features, target, model, 10)
```

### 5. Extend the analysis and cross-validation to bank-additional-full.csv. How does the performance change?

```python
data_full = pd.read_table('bank-additional-full.csv', sep=';')
data_full.head()
```

```python
data_full_clean = data_full

new_label_vars = ["marital", "job", "education", "contact", "month", "poutcome", "default", "housing", "loan", "y", "day_of_week"]

for i in new_label_vars:
    le = preprocessing.LabelEncoder()
    le.fit(data_full_clean[i].unique())
    newvarname = "new_" + str(i)
    data_full_clean[newvarname] = le.transform(data_full_clean[i]).tolist()

data_full_clean.drop(new_label_vars, inplace=True, axis=1)
```

```python
feature_names = data_full_clean.columns.tolist()
feature_names.remove("new_y")

features = data_full_clean[feature_names].values
target = data_full_clean['new_y'].values
```

```python
model = tree.DecisionTreeClassifier(random_state=0).fit
cross_validate(features, target, model, 10)
```

```python
model_rf = RandomForestClassifier().fit
cross_validate(features, target, model_rf, 10)
```

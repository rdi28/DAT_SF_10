We begin by importing packages that we will be using: pandas, matplotlib, numpy, and statsmodels (for regression analysis). We use matplotlib inline to set up plotting in the ipython notebook.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm

%matplotlib inline
```

```python
data = pd.read_csv('student_logins.csv')

data.head()
```

Change session_start_time, session_end_time, and student_account_created to be day time fields

```python
data['session_start_time']=pd.to_datetime(data['session_start_time'])
data['session_end_time']=pd.to_datetime(data['session_end_time'])
data['student_account_created']=pd.to_datetime(data['student_account_created'])
```

Add a column for session_length by subtracting session_start_time from session_end_time

```python
data['session_length']=data['session_end_time']-data['session_start_time']
```

Add a column for age of session by subtraction session_start_time from account created at time

```python
data['age_of_account']=data['session_start_time']-data['student_account_created']
```

Start by reading the data into a pandas dataframe and taking a look at what we have. 

Look at data, sorted by length of session - the variable we are trying to predict.

```python
result = data.sort('session_length')
result.head()
```

There's some bad data, with a session_end_time before a session_start_time. This will likely affect the overall model, so we want to eliminate the data.

```python
data_1 = data.drop(38511)
```

```python
result = data_1.sort('student_id')
result.head()
```

Take session length & age of account and convert into seconds and hours respectively

```python
data_1['session_length'] = data_1['session_length'].astype('timedelta64[s]')
data_1['age_of_account'] = data_1['age_of_account'].astype('timedelta64[D]')
data_1.head()
```

Check what the scatter matrix looks like

```python
from pandas.tools.plotting import scatter_matrix
scat = scatter_matrix(data_1, figsize = (8,8))
```

Drop uneeded data.

```python
data_2 = data_1.drop(['session_start_time', 'session_end_time', 'student_account_created'], 1)
data_2.head()
```

Create a dictionary for class_id and change it into a numerical field. 

```python
labels_dict = {name : i for i, name in list(enumerate(np.unique(data_2['class_id']))) }

labels_dict
```

Map class_id to the numerical field.

```python
data_2['class_id_mapped'] = data_2['class_id'].map(lambda x: labels_dict[x])

data_2.head()
```

Run the regression

```python
x = data_2[['problems_completed', 'student_previous_logins_total', 'student_previous_class_logins', 'student_id', 'class_id_mapped', 'age_of_account']].values
y = data_2[['session_length']].values
```

```python
X = sm.add_constant(x, prepend=True)
results = sm.OLS(y,X).fit()
```

```python
results.summary()
```

R squared of 0.415 is not great lets see if we can do any better. Lets add some dummy variables for the class id instead.

```python
for j in data_2['class_id'].unique():
    label = "class_" + str(j)
    data_2[label] = data_2['class_id'] == j
```

```python
data_2.head()
```

New model.

```python
x1 = data_2[['problems_completed', 'student_previous_logins_total', 'student_previous_class_logins', 'student_id', 'age_of_account', 'class_e', 'class_m', 'class_g', 'class_c', 'class_a']].values
y1 = data_2[['session_length']].values
```
